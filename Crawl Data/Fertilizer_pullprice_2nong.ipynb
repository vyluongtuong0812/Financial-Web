```python
import requests

def get_all_products():
    # Gửi yêu cầu HTTP để lấy tất cả sản phẩm
    url = "https://api-production.2nong.vn/v0/products?page=1&is_active=1&is_favorite=&sort[updated_at]=desc&special_id=2"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json().get('data', [])  # Lấy ra phần data từ response JSON, nếu không có thì trả về list rỗng
    else:
        return None

def get_product_price(products):
    for product in products:
        product_id = product.get('id')
        city_id = product.get('city_id')
        price = product.get('price')
        product_prices.append((product_id, city_id, price))
    return product_prices

# Lấy tất cả sản phẩm
all_products = get_all_products()

# Lấy giá cho tất cả các sản phẩm và thành phố
product_prices = get_product_price(all_products)

# In ra thông tin của từng sản phẩm
for product in all_products:
    product_id = product['id']
    city_id = product['city_id']
    price = product['price']

    print(f"Giá phân bón {product_id} của {city_id}: {price}")
```

```python
import pandas as pd 
```

```python
response = requests.get("https://api-production.2nong.vn/v0/products?page=1&is_active=1&is_favorite=&sort[updated_at]=desc&special_id=2")

# Xử lý dữ liệu JSON
data = response.json()
```

```python
df_init = pd.DataFrame(data['data'])[['id','name','unit','special_id','is_active','city_id','city_name','created_at','updated_at']]
df = pd.DataFrame(data['data'])[['id','info']]

```

```python
info_convert_list = []

for i in range(len(df)):
    info_value = df.loc[i, 'info']
    info_df = pd.DataFrame(info_value)
    info_df.columns = info_df.columns.str.lower()

    # Ghép nối với cột id, sử dụng id từ hàng hiện tại cho tất cả các hàng trong info_df
    info_convert_df = pd.concat([pd.Series([df.loc[i, 'id'] for _ in range(len(info_df))]), info_df], axis=1, names=['id', None])

    info_convert_list.append(info_convert_df)

info_convert = pd.concat(info_convert_list, ignore_index=True)
# Đặt tên trục cột, bao gồm 'id'
info_convert.rename(columns={0: 'id','datetime':'act_date'}, inplace=True)
```

```python
df_merged = pd.merge(left=df_init, right=info_convert, on=['id','city_id'], how='outer')
df_merged
```
